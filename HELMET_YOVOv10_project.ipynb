```python
YAML_PATH = '/content/yolov10/Safety_Helmet_Dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 32

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            batch=BATCH_SIZE,
            imgsz=IMG_SIZE)
```

```python
#evaluate model
TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train2/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```

```python
#run prediction with an image
# https://github.com/googlecolab/colabtools/issues/3409
import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
from google.colab.patches import cv2_imshow

TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train2/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

IMAGE_URL = 'https://gro.com.vn/wp-content/uploads/2021/03/non-bao-hiem-tre-em.jpg'
CONF_THRESHOLD = 0.3
results = model.predict(source=IMAGE_URL,
                       imgsz=IMG_SIZE,
                       conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
```
